## Part 3: Reflection Paper

### 1. Purpose and Functionality

The WGU medical readmission dataset contains various variables that pertain to a sample of patients that have entered the hospital network. The dataset was originally created to look into causes of readmission however we do not have to examine readmission and so I decided to focus on costs. As a nationwide hospital network it's important for, what I have aliased, the WGU Hospital Network, to know how each state they are in perform in comparison to their competitor hospitals. While looking at hospital costs as a whole is a worthwhile metric, not all states have equal average hospital costs and therefore looking at where WGU Hospital Network differs from other hospitals will help adjust performance and allow the WGU Hospitals to be competitive in their local areas. 

With that in mind I created a dashboard that will allow the stakeholders of the WGU Hospital Network to compare the cost averages of the WGU hospitals across the states using various of the variables included as metrics to help explain where costs differ. Since this analysis is less focused on individual health and outcomes I focused on the variables that are more general. These include the cost, admission type, days spend in hospital, and complication risk, while maintaining the state variable as the base variable. 

### 2. Variables and Insights

I used the Healthcare Cost and Utilitization Project (HCUP) public data to compare the WGU Hospital costs to other hospitals in the United States. The HCUP gathers healthcare databases to analyze the healthcare costs, insurance costs, and other healthcare related issues. Since the WGU data is undated I decided to use the State Inpation Database from 2019 since it is recent enough but pre-pandemic when medical data gets less typical. 

The 2019 Inpatient Dataset in fairly simple with total cost, expected payer, state, and what type of initial admission. Since I am mostly interested in how the WGU Hospital Network compares by state cost this data was a perfect compliment and allowed for interesting insights into where the WGU Hospital Network differs. The variables for expected payer and initial admission are much more detailed than the WGU dataset which does not allow for one to one comparison but does help to illustrate how breaking data into more specific values helps the analysis overall. I think this is also an important insight to bring to the WGU Hospital Network since the dataset provided is generally too vague for better insights so illustrating how specificity of data gathering enhances analysis and leads to better decision making will help to persuade the stakeholders to invest in a better data gathering initiative.

### 3. Two different Data Representations

One data representation that I created is a map of the united states where I have used color as a way to represent the cost of hospitalization in each state. I duplicated that map for both the WGU Hospital Network dataset and the HCUP dataset. Since the cost variables are on very different scales for the two datasets I elected to illustrate the cost by percentile. Having the two maps beside each other will allow the stakeholders to examine where WGU Hospital costs differ from the average. For example when looking at the map the most obvious difference is Alaska where the HCUP data shows Alaska being one of the cheaper hospitalization states and the WGU data shows it being one of the most exensive. This obvious difference will allow stakeholders to see that doing further data gathering and analysis focusing on Alaska would be a good next step to ensure that the WGU Hospitals in Alaska are staying competitive. Likewise the data from Florida shows that WGU costs in Florida are lower than the HCUP averages and again the stakeholders might want to put funds into research as to why Florida is cheaper for WGU Hospitals than other hospitals. 

While comparing the costs from states is the first objective of this project, the second objective is illustrating to the stakeholders why better data gathering inititives are needed to actually solve some of the issues of differing costs. To do this I creaed a bar chart which shows the average length of hospital stay in each state in the WGU Hospital Network and mapped on the percentile of average cost via color. This chart clearly shows that the cost of hospital is directly correlated to the length of stay. Which might make the stakeholders ask what effects the length of stay? To look into that I created two charts that show the cost of each state with filters for length of stay and color coding for both services and complication risk, both of these charts illustrate to stakeholders that the current data that we have for the WGU Hospital System does not effectively give us the answers to make any business decisions about how to normalize costs in states that we differ in. Therefore the best course of action would be to rewrite a data collecting initative to roll out in the states that we have previously singled.

### 4. Two Interactive controls

Each of the visualizations I created has many different controls to filter down the data so that stakeholders can see as granular of information as they choose. One of these interactive controls is filtering states from a multiple choose drop down menu. This filter allows the users to filter down each visualization to just focus on the state or states that they want to look at specific metrics for.

A second filter I have is to filter by the cost. This slider allows users to filter out states by their cost so they can see the highest cost states, lowest cost, or some middle cost that they choose. In conjunction these two filters can allow users to really drill down to specific results. 

### 5. Building and Accessibility

In order to create a dashboard that is accessible I chose colorblind friendly color schemes, using both this incredible tool by Venngage and Tableau's own colorblind color scheme. I also made sure that none of the charts solely used color to impart information. All of the charts have tooltips that also include the information.

### 6. Two data respresentations

The story I wanted to tell was how the specificity of the HCUP data leads to better data analysis than the WGU dataset. I did this first through the visualizations I created that break down the cost of hospitalization by payer for each state. In the WGU data I only had age of patient so I broke it into two broad parts, 18-64 and 65+ to simulate the medicare transition. However in the chart of HCUP data I was able to use all of the payer categories and by comparing the two charts it is easy to see how the breakdown of payer type really assists in the analysis of cost of hospitalization which will lead to better data driven business decisions. 

The second visualizaiton that works to illustrate the benefit of specificity in data collection is the chart I created for the breakdown of type of hospitalization. Within the WGU data the chart for admission type is incredibly uniform which makes it hard to make any analysis of while the HCUP data has many more types of admission which allows for much easier analysis. For instance it is easy to see for stakeholders that the highest type of admission is Adult Hospitalization while injury is fairly small but persistent. 

### 7. Audience Analysis

The audience that I was presenting to were the stakeholders for the WGU Hospital Network, therefore people like the President and Vice President of Research and Hospital Operations as well as the board of directors. As these people are not fellow data analysts I wasn't interesting in presenting the gritty details of the data. Instead I wanted to present a high concept comparison of the datasets and illustrate the ways that the data differs. Similarly I wanted to present the data in a way that showed why better data needed to be collected in order to create the sort of business decisions that the hospital network would need to continue to be successful. 

### 8. Universal Access

In order to have universal access to the dashboard as well as no barrier for that access I hosted my dashboard on Tableau Public so that anyone with the link will be able to use the dashboard. 

### 9. Two Elements of Storytelling

Using the elements of storytelling that were taught in the Datacamp courses I started my presentation with broad data presentation and then transistioned into the more segmented data. This way the audience was able to get a feel for the large scope of the data analysis before being presented with more granular concepts. Another aspect of storytelling that I used was to always return to the conclusion of the analysis with each explanation. That way the audience remembered and could track how I came to that conclusion as I presented the analysis. Together this way of data storytelling kept my presentation focused and easy to follow for the audience as well as keeping the question and conclusion in mind.

## Data Cleaning of the WGU Data Set 

This dataset cleaning code comes from my previous projects.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno
import seaborn as sb
%matplotlib inline

#Import the dataset and make sure the bonus column is not created.
df = pd.read_csv('medical_clean.csv', index_col=0)
# Check dataset
df.info()
```

```python
#Changing Datatypes
#Convert area to category from string
df["Area"] = df["Area"].astype("category")
# Convert column to category from string
df["TimeZone"] = df["TimeZone"].astype("category")
# Convert column to category from string
df["Marital"] = df["Marital"].astype("category")
# Convert column to category from string
df["Gender"] = df["Gender"].astype("category")
# Convert column to category from string
df["Initial_admin"] = df["Initial_admin"].astype("category")
# Convert column to category from string
df["Complication_risk"] = df["Complication_risk"].astype("category")
# Convert column to category from string
df["Services"] = df["Services"].astype("category")
```
